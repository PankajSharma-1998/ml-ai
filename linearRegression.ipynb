```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
# read data using panda

df = pd.read_csv('/content/sample_data/homeprices.csv') #pd.DataFrame({'area': area, 'price': price})
new_def = df.drop('price', axis='columns')
df
```

```python
# plot chart
%matplotlib inline
plt.scatter(df.area, df.price, marker='*')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
# apply linear model
# as this is the supervised learning

reg = linear_model.LinearRegression()

# fit area against price
reg.fit(new_def, df.price)
```

```python
#  now we are ready to predict prices
def predict(areas):
  return np.add(reg.coef_ * np.array(areas), reg.intercept_)

# predicting price for test area, it is showing linear model is fit for prediction
predicted_output = predict(df.area)

# predicting price for train data
# predicted_output = predict([2000, 5000, 6000, 7000, 1000])

#reg.predict(np.array([[3000]]))

predicted_output
```

```python
# plot regression line

plt.ylabel('area')
plt.xlabel('price')
plt.title('area-price chart')
plt.scatter(df.area, df.price)
plt.plot(df.area, predicted_output, color='red')
plt.show()
```
